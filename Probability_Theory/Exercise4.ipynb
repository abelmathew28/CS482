```python
# Import the required libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# This is the function to generate the two correlated random variables
# Input: Required Correlation between random variables
# Output: Random variables having correlation as required

def generate_corelated_random_variables(corr):
  theta = np.arcsin(corr)/2

  a = d = np.cos(theta)
  b = c = np.sin(theta)

  Y1 = np.random.normal(0,1, 1000)
  Y2 = np.random.normal(0,1, 1000)

  X1 = a*Y1 + b*Y2
  X2 = c*Y1 + d*Y2

  return X1, X2
```

```python
# This code is to generate the subplot
# Input: ax - defines the axis for the subplot; corr - defines the required correlation between random variables; pos - defines the position of the subplot
 
def plotnew(ax, corr, pos):
  X1, X2 = generate_corelated_random_variables(corr)
  ax.scatter(X1, X2, color=['black'])
  ax.set_xlabel("X1")
  if (pos == 1) | (pos==3):
    ax.set_ylabel("X2")
  else:
    ax.set_yticklabels([])
  ax.set_xlim([-4.8, 4.8])
  ax.set_ylim([-4.8, 4.8])
  ax.text(3, 4,f'p = {corr}', horizontalalignment='center', verticalalignment='center')
```

```python
# This code creates the 4 subplot and display each subplot for random variables having requested correlation

plt.rcParams["figure.figsize"] = (10,10)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2)
plotnew(ax1,  0.00, 1)
plotnew(ax2,  0.20, 2)
plotnew(ax3, -0.80, 3)
plotnew(ax4,  0.99, 4)

fig.suptitle("Correlated GRVs from N(0,1) Random Numbers")
plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```
